```python
import numpy as np
import matplotlib.pyplot as plt

# Load FID scores from the specified files
fid_dcgan = np.load('./Results/FID_scores/DCGAN_FID.npy')
fid_wgangp = np.load('./Results/FID_score/WGANGP_FID.npy')
fid_acgan = np.load('./Results/FID_scores/ACGAN_FID.npy')
```

```python
# Create a plot for FID score comparisons
plt.figure(figsize=(10, 5))
plt.title("Comparison of FID Scores: DCGAN, WGAN, and ACGAN")
plt.plot(fid_dcgan, label="DCGAN", color='red')
plt.plot(fid_wgangp, label="WGAN", color='green')
plt.plot(fid_acgan, label="ACGAN")
plt.xlabel("Epochs")
plt.ylabel("FID Score")
plt.legend()
```

```python
# Save the plot as a JPEG file
plt.savefig('./Results/FID_summary.jpg', format='jpeg', dpi=100, bbox_inches='tight')
plt.show()
```
